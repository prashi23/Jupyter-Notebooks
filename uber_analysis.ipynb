## Loading libraries

```python
%pylab inline
import pandas
import matplotlib
import seaborn
```

## Load CSV file 

```python
data = pandas.read_csv('uber-raw-data-apr14.csv')
```

```python
data
```

```python
data['Date/Time']
```

```python
data.tail()
```

```python
data.head()
```

```python
data['Date/Time'] = data['Date/Time'].map(pandas.to_datetime)
```

```python
data.tail()
```

```python
def get_dom(dt):
    return dt.day

data['dom'] = data['Date/Time'].map(get_dom)
```

```python
data.tail()
```

```python
def get_weekday(dt):
    return dt.weekday()

data['weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour

data['hour'] = data['Date/Time'].map(get_hour)

data.tail()
    
```

## Analysis

### ANALYSE THE DOM

```python
hist(data.dom, bins=30, rwidth = 0.8, range = (0.5, 30.5) )
xlabel('date of the month')
ylabel('frequency')
title('Frequency by DoM - UBER ANALYSIS - Apr2014 ')
```

```python
for k, rows in data.groupby('dom'):
    print(k, len(rows))
    
def count_rows(rows):
    return len(rows)

by_date = data.groupby('dom').apply(count_rows)
by_date

```

```python
by_date_sorted = by_date.sort_values()
```

```python
bar(range(1, 31), by_date_sorted)
xlabel('date of the month')
ylabel('frequency')
title('Frequency by DoM ')
```

```python
by_cross = data.groupby('weekday hour'.split()).apply(count_rows).unstack()
```
