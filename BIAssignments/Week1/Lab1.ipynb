```python
# Q1.Write a python program (menu driven) that asks for an operation to be performed on two input numbers. 
# Operations are add, subtract, divide, multiply. 
# The program should keep asking for options on new numbers until user specifies explicitly to exit. 
# Also handle the scenario of divide by zero through exception handling in python.

#all the addition, subtraction, multiplication and other operations in functions 
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y
    
def divide(x, y):
    try:
        print(f"{x} / {y} is {x/y}")

    except ZeroDivisionError:
        print("Cannot divide by zero")
    

# input the operands 
num1 = float(input("Enter the first number"))
num2 = float(input("Enter the second number"))

# printing the operations
print("Available operations")
print("1.Addition")
print("2.Subtraction")
print("3.Mutiplication")
print("4.Divide")

#taking the input from user to select the operation that he/she wants to perform
choice = input("Which operation you wanna perform? ") 

if choice == '1':
    print(num1,"+",num2,"=", add(num1,num2))
elif choice == '2':
    print(num1,"-",num2,"=", subtract(num1,num2))
elif choice == '3':
    print(num1,"*",num2,"=", multiply(num1,num2))
elif choice == '4':
    print(num1,"/",num2,"=", divide(num1,num2))
else:
    print("Invalid input") 
```

```python
#Q2.Write a python program that takes a sentence as input from user and reverses the words in the sentence. 
#Eg. If the sentence is 'Welcome to AI lab', then the output should be 'lab AI to Welcome'.

# break the words of a string. After that reverse this list and join them.
#s.split(' ') splits the strings using space as a separating character. It returns an array of words that have been split.
#reversed(list) reverses the string.
#'separator'.join(words) joins the words using the separator that is space here. 

# But this reverses the punctuations.
s = input("Enter the sentence that you want to reverse")
r = ' '.join(reversed(s.split(' ')))
print(r)
```

```python
#Another solution to the second question 
#having fun with regular expressions

#punctuations is an array of all punctuations.
#search any punctuations till end of the line so we can add it back to the result string.
#The $ symbol in regex to specify that the search is for the end of line. regex used to search all punctuations till the end of line and store them in a string variable.
#strip the last punctuations and then split rest of the words. After that reverse that array and join them.
#and then add back the end punctuations.

import re

punctuations = ':;,.!?'
s = input("Enter the sentence that you wanna reverse")
match = re.search(r'([:;,.!?]+)$', s)
end_puncs = match.group(1) if match is not None else ''
rev = ' '.join(reversed(re.findall('[\w:;,.!?]+', s.rstrip(punctuations)))) + end_puncs
print(rev)
```

```python
#q4.Write a python program to find square root of a given number (n) within a given precision (p). 
#Do not use any predefined library. 
def sRoot(number, precision):
    start = 0
    end = number//2
 # For computing integral part
    # of square root of number
    while (start <= end) :
        mid = int((start + end) / 2)
         
        if (mid * mid == number) :
            ans = mid
            break
          # incrementing start if integral
        # part lies on right side of the mid
        if (mid * mid < number) :
            start = mid + 1
            ans = mid
         # decrementing end if integral part
        # lies on the left side of the mid
        else :
            end = mid - 1
         # For computing the fractional part
    # of square root upto given precision
    increment = 0.1
    for i in range(0, precision): 
        while (ans * ans <= number):
            ans += increment
         # loop terminates when ans * ans > number
        ans = ans - increment
        increment = increment / 10
    return ans
# Driver code
n=int(input("Enter number whose square root is to be calculated "))
p=int(input("Enter the precision "))
print("SQUARE ROOT WITH PRECISION IS "+str(round(sRoot(n, p), p)))
```

```python
#q5.Find the minimum value of a function y = (x+3)^2, where ^ means power. 
#During lab evaluation, demonstrate the solution to this problem using hand-written plot. 
#Write python program to find the minimum value programmatically, use gradient descent algorithm. 

cur_x = 3 # The algorithm starts at x=3
rate = 0.01 # Learning rate
precision = 0.000001 #This tells us when to stop the algorithm
previous_step_size = 1 #
max_iters = 10000 # maximum number of iterations
iters = 0 #iteration counter
df = lambda x: 2*(x+3) #Gradient of our function

while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x #Store current x value in prev_x
    cur_x = cur_x - rate * df(prev_x) #Grad descent
    previous_step_size = abs(cur_x - prev_x) #Change in x
    iters = iters+1 #iteration count
    print("Iteration",iters,"\nX value is",cur_x) #Print iterations
    
print("The local minimum occurs at", cur_x)
```

```python
#q6.Create a suitable class in python to represent the mathematical concept of 'set'. 
#Create appropriate member variables and member functions of this class to perform operations: Union, Intersection, Cardinality.

class set:
    def abc(self, setA, setB):
        self.setA = setA
        self.setB = setB
    def unionFunc(self):
        return self.setA.union(self.setB)
    def intersectionFunc(self):
        return self.setA.intersection(self.setB)
    def cardinalityFunc(self, setName):
        return len(setName)
setObj = set()
setA = {1,2,3,4}
setB = {2,4,6,8}
setObj.abc(setA, setB)
setUnion = setObj.unionFunc()
setIntersection= setObj.intersectionFunc()
setCardinality= setObj.cardinalityFunc(setA)
print(setUnion)
print(setIntersection)
print(setCardinality)
```
