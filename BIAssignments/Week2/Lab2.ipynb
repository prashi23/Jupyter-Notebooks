```python
#Scenario: Download the dataset and understand the various attributes of the dataset.
# https://github.com/fivethirtyeight/data/blob/master/us-weather-history/KCLT.csv
#q1.1. Load the csv file of dataset into dataframe in Python program.

import numpy 
import pandas
import matplotlib
data = pandas.read_csv('KCLT.csv')
data
```

```python
#2. Find the mean, median, mode, min and max for all numeric attributes.

for i in data.columns[1:]:
    print(i, ":", data[i].mean())

for i in data.columns[1:]:
    print(i, ":", data[i].median())

for i in data.columns[1:]:
    print(i, ":", data[i].mode())
    
for i in data.columns[1:]:
    print(i, ":", data[i].min())
    
for i in data.columns[1:]:
    print(i, ":", data[i].max())
```

```python
#3. Plot min and max for actual_mean_temp, actual_min_temp and actual_max_temp using bar graph. 
# Refer: https://pythonspot.com/matplotlib-bar-chart/ (last example).

import matplotlib.pyplot as plt
import numpy as np

#data to be plot
n_groups = 3
minimums = (min(data['actual_mean_temp']),min(data['actual_min_temp']),min(data['actual_max_temp']))
maximums = (max(data['actual_mean_temp']),max(data['actual_min_temp']),max(data['actual_max_temp']))
#create a plot
# subplots are to plot two or more plots in one figure
fig,ax = plt.subplots() 
index = np.arange(n_groups)
bar_width = 0.5
opacity = 0.6

#plt.bar(x,height,bar_width)
result1 = plt.bar(index,minimums,bar_width,alpha=opacity,color='b',label='Minimum')
result2 = plt.bar(index+bar_width,maximums,bar_width,alpha=opacity,color='r',label='Maximum')

plt.xlabel('Categories of Temperature')
plt.ylabel('Values')
plt.title('Comparison of min, max and mean')

#xticks to arrange 
#xticks(ticks, [labels], **kwargs)  # Set locations and labels
plt.xticks(index+bar_width,('Actual_mean_temp','Actual_min_temp','Actual_max_temp'))

plt.legend()
#for showing what different colors represent

plt.tight_layout()  #adjust subplot params so that subplots are nicely fit in the figure
plt.show() 
```

```python
#4. Plot two separate histograms for actual_mean_temp for months of Aug 2014 and March 2015.
# Make key observations (put them as comments in jupyter notebook). 
# Refer: https://matplotlib.org/1.2.1/examples/pylab_examples/histogram_demo.html
import numpy 
import datetime
import pandas as pd

df2 = pd.read_csv("KCLT.csv",parse_dates=['date'])

cols = [col for col in df2.columns if col in ['date','actual_mean_temp']] 
df3 = df2[cols]

df4=df3[(df3['date']>pd.Timestamp(2014,7,31)) & (df3['date']<pd.Timestamp(2014,9,1))]

#plt.hist(x, bins=None, range=None, density=None)
df4.hist(bins=15)
plt.xlabel('Mean Temperature')
plt.ylabel('Probability')
plt.title('Actual Mean Temperature for Aug,2014')
plt.grid(True)

#For March 2015
df4=df3[(df3['date']>pd.Timestamp(2015,2,28)) & (df3['date']<pd.Timestamp(2015,4,1))] #for keeping relevant dates

#plt.hist(x, bins=None, range=None, density=None)# x should be an array
df4.hist(bins=15)
plt.xlabel('Mean Temperature')
plt.ylabel('Probability')
plt.title('Actual Mean Temperature for March,2015')
plt.grid(True)
```

```python
#5. Plot box plots for attributes actual_mean_temp, actual_min_temp and actual_max_temp. 
# Make key observations (put them as comments in jupyter notebook). 
# Refer:https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.boxplot.html

cols = [col for col in data.columns if col in ['actual_mean_temp', 'actual_min_temp', 'actual_max_temp']]
df2=data[cols]

boxplot = df2.boxplot(column=['actual_mean_temp', 'actual_min_temp', 'actual_max_temp'])
```

```python
#6. Find the correlation between the following pairs of attributes: actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. 
#Make key observations.

# Between actual_mean_temp & actual_min_temp
#to find the pairwise correlation of all columns in the dataframe.
print(data["actual_mean_temp"].corr(data["actual_min_temp"]))

                 
# Between actual_mean_temp & actual_max_temp
print(data["actual_mean_temp"].corr(data["actual_max_temp"]))

#They are highly correlated. Both the quantities are almost equally correlated to the mean
```

```python
#7. Plot two scatter plots between the following pairs of attributes: actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. 
#Make key observations. 
#Refer: https://matplotlib.org/gallery/shapes_and_collections/scatter.html#sphx-glr-gallery-shapes-and-collections-scatter-py

data.plot.scatter("actual_mean_temp", "actual_min_temp")
plt.title("Scatter plot b/w actual_mean_temp & actual_min_temp")

data.plot.scatter("actual_mean_temp", "actual_max_temp")
plt.title("Scatter plot b/w actual_mean_temp & actual_max_temp")
```

```python
#8. Make the correlation matrix plot and scatterplot matrix taking all temperature attributes in the dataset. 
#Refer:https://machinelearningmastery.com/visualize-machine-learning-data-python-pandas/


#Correlation gives an indication of how related the changes are between two variables.
#correlation is calculated between every pair of variables.
#This is called a correlation matrix
#scatter matrix contains for each combination of the variable , the relation between them 
correlations = data.corr()
fig = plt.figure() #Create a new figure
ax = fig.add_subplot(111) #"111" means "1x1 grid
cax = ax.matshow(correlations, vmin=-1, vmax=1) #Display an array as a matrix in a new figure window.
#min and vmax define the data range that the colormap covers. By default, the colormap covers the complete value range of the supplied data. 
fig.colorbar(cax) #colorbar for different colors
ticks = np.arange(0,12,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_yticklabels(data.columns)
plt.show()


#ScatterPLot
from pandas.plotting import scatter_matrix
scatter_matrix(data, figsize = [20, 20])
plt.show()
```
